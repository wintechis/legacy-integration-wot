```python
import os

folder_path = './data/product_documentations'
file_list = []

for root, dirs, files in os.walk(folder_path):
    for file in files:
        file_path = os.path.join(root, file)
        
        file_list.append(file_path)

print(file_list)
```

```python
import re
file_dict = {}

for file_path in file_list:
    folder = re.search(r'(?<=product_documentations\/).*', file_path)

    
    if folder:
        folder = folder.group().split('/')
        folder_name = folder[0]
        file_path = folder[1]
            
        if folder_name in file_dict:
            file_dict[folder_name].append(file_path)
        else:
            file_dict[folder_name] = [file_path]

print(file_dict)
```
